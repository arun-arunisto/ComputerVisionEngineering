## Converting MRI Scans to Colored Images using CV2, Matplotlib, MONAI (Medical Open Network for Artificial Intelligence)

***Using CV2***

```python
import cv2
```

```python
#loading the image to convert
mri_image = cv2.imread('/content/drive/MyDrive/data/brain_tumor_dataset/train/meningioma/0000.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mri_image, cmap='gray')
```

## applying Color map to the mri image

```python
colored_mri = cv2.applyColorMap(mri_image, cv2.COLORMAP_JET)
plt.imshow(colored_mri[:, :, ::-1])
```

## Using Matplotlib

```python
import matplotlib.cm as cm
```

```python
mri_image_2 = cv2.imread("/content/drive/MyDrive/data/brain_tumor_dataset/train/meningioma/0001.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(mri_image_2, cmap='gray')
```

```python
# Normalizing the image to [0, 1] for applying colormaps
mri_image_normalized = mri_image_2 / 255.0
```

```python
# Applying  colormap (e.g., 'jet')
colored_mri = cm.jet(mri_image_normalized)
```

```python
plt.imshow(colored_mri)
```

```python
import monai
from monai.transforms import ScaleIntensity
import numpy as np
```

```python
#loading the image to convert
mri_image_3 = cv2.imread("/content/drive/MyDrive/data/brain_tumor_dataset/train/meningioma/0002.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(mri_image_3, cmap='gray')
```

```python
# Applying MONAI transformation for intensity scaling
transform = ScaleIntensity(minv=0.0, maxv=1.0)
mri_image_scaled = transform(mri_image_3)
```

```python
# Applying colormap using Matplotlib
colored_mri = cm.viridis(mri_image_scaled)
```

```python
plt.imshow(colored_mri)
```
