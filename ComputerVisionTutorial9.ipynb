#Creating Panoramas using OpenCV

```python
#import libraries
import cv2 as cv
import math
import numpy as np
import matplotlib.pyplot as plt
```

##Steps for creating panoramas
1. Finding Keypoints of all images
2. finding pairwise correspondence
3. estimating pairwise homographies
4. refine homographies
5. stitch with blending

```python
#reading images
imagefiles = ["boat1.jpg", "boat2.jpg", "boat3.jpg",
              "boat4.jpg", "boat5.jpg", "boat6.jpg"]
```

```python
images = []
for filename in imagefiles:
  img = cv.imread(f"/content/drive/MyDrive/ComputerVisionTutorial/boat/{filename}")
  img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  images.append(img)
num_images = len(images)
```

```python
#displaying images
plt.figure(figsize=[30, 10])
num_cols = 3
num_rows = math.ceil(num_images/num_cols)
for i in range(0, num_images):
  plt.subplot(num_rows, num_cols, i+1)
  plt.axis("off")
  plt.imshow(images[i])
```

##we are going to stitch the images using sticher class

```python
#stitching images
stitcher = cv.Stitcher_create()
status, result = stitcher.stitch(images)
if status == 0:
  plt.figure(figsize=[30, 10])
  plt.axis("off")
  plt.imshow(result)
  plt.title("Panorama")
```
