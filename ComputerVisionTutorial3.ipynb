<h3>Annotating Images</h3>
<p>Annotating images mean <br>
<ul>
<li>Draw lines</li>
<li>Draw circles</li>
<li>Draw rectangles</li>
<li>Adding text</li>
</ul></p>

```python
#importing required packges
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading an image
falcon = cv.imread("/content/drive/MyDrive/ComputerVisionTutorial/ImagesToPractise/falcon9.jpg")
falcon_rgb = cv.cvtColor(falcon, cv.COLOR_BGR2RGB)
plt.imshow(falcon_rgb)
```

<h2>Drawing a line</h2>
<p>for drawing a line we are going to use <b>cv2.line()</b> function</p>
<h4>Function syntax</h4>
<p>
<b>img = cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])</b>
<br>mainly, function required four arguments<br>
<ol>
<li><b>img</b> : image pn which we draw line</li>
<li><b>pt1</b> : first point(x, y location) of the segment</li>
<li><b>pt2</b> : second point of the line segment</li>
<li><b>color</b> : color of the line which will be drawn</li></ol>
other optional arguments.
<ol>
<li><b>thickness</b> : specifying line thickness default value is 1</li>
<li><b>lineType</b> : default value is 8-connected line. cv2.LINE_AA (antialiased or smooth line)</li></ol></p>

```python
falcon_line = falcon_rgb.copy() #copying image
cv.line(falcon_line, (700, 100), (900, 100), (255, 0, 255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(falcon_line)
```

<h3>Drawing a circle</h3>
<p>we are going to draw circle using <b>cv2.circle()</b> function</p>
<h4>Function Syntax</h4>
<p><b>img = cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])</b><br>
<b>The function required 4 arguments</b><br>
<ol>
<li><b>img</b> : image on which we want draw circle.</li>
<li><b>center</b> : center of the circle</li>
<li><b>radius</b> : radius of the circle</li>
<li><b>color</b> : color of the circle</li></ol>
<b>Optional arguments</b><br>
<ol>
<li><b>thickness</b> : thickness of the circle</li>
<li><b>lineType</b> : line type of the circle</li></ol></p>

```python
#drawing circle
falcon_circle = falcon_rgb.copy()

cv.circle(falcon_circle, (590, 400), 100, (0, 0, 255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(falcon_circle)
```

<h3>Drawing rectangle</h3>
<p>to draw a rectangle we are using <b>cv2.rectangle()</b> function</p>
<h4>Function Syntax</h4>
<p><b>img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])</b>
<br><b>the function required 4 arguments</b><ol>
<li><b>img</b> : source image to draw the rectangle</li>
<li><b>pt1</b> : top-left vertex</li>
<li><b>pt2</b> : bottom-right vertex</li>
<li><b>color</b> : rectangle color</li></ol>
<b>optional Arguments</b>
<ol>
<li><b>thickness</b> : thickness of the outline, if a negative value applied it will fill rectangle</li>
<li><b>lineType</b> : cv2.line default</li>
</ol></p>

```python
#drawing a rectangle
falcon_rectangle = falcon_rgb.copy()

cv.rectangle(falcon_rectangle, (700, 100), (900, 400), (255, 0, 255), thickness=5, lineType=cv.LINE_8)
plt.imshow(falcon_rectangle)
```

<h3>Adding Text</h3>
<p>to add a text we will use <b>cv2.putText()</b> function</p>
<h4>Function syntax</h4>
<p><b>img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])</b>
<br>
<b>function required 6 arguments</b>
<ol>
<li><b>img</b> : image where text has to be written</li>
<li><b>text</b> : text string to be written</li>
<li><b>org</b> : bottom-left corner of text string in the image</li>
<li><b>fontFace</b> : font type</li>
<li><b>fontScale</b> : font-size</li>
<li><b>color</b> : font color</li></ol>
<b>Optional arguments</b>
<br>
<ol>
<li><b>thickness</b> : thickness of the text</li>
<li><b>lineType</b> : default value is 8 usually we use cv2.LINE_AA</li></ol></p>

```python
#adding text to image
falcon_text = falcon_rgb.copy()
text = "Space X Falcon 9 ->"
fontScale = 3
fontFace = cv.FONT_HERSHEY_PLAIN
fontColor = (0, 0, 0)
fontThickness = 2
cv.putText(falcon_text, text, (180, 200), fontFace, fontScale, fontColor, fontThickness, cv.LINE_AA)
plt.imshow(falcon_text)
```
